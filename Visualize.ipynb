```python
# Visualize
import matplotlib.pyplot as plt
def visualize(X,y,colors):
    plt.scatter(X, y, c=colors)
# Set the title and labels
    plt.title('Scatter Plot')
    plt.xlabel('Index')
    plt.ylabel('Angle')
    plt.show()


```

```python
# Head

y = dataset[3].values
X = list(range(len(y)))

labels = dataset[0].values
colors = []
for l in labels:
    if l == 0:
        colors.append('red')
    elif l == 1:
        colors.append('blue')
visualize(X, y,colors)
```

```python
# Back

y = dataset[4].values
X = list(range(len(y)))

labels = dataset[1].values
colors = []
for l in labels:
    if l == 1:
        colors.append('red')
    elif l == 2:
        colors.append('blue')
    elif l == 3:
        colors.append('green')
visualize(X, y,colors)
```

```python
# Left right

y = dataset[5].values
X = list(range(len(y)))

labels = dataset[2].values
colors = []
for l in labels:
    if l == 0:
        colors.append('red')
    elif l == 1:
        colors.append('blue')
visualize(X, y,colors)
```
